## 3 transformación de columna


La idea para este dataset predecir la calidad del vino, pero, como mencionado anteriormente, tenemos una columna que no es numerica, así que no la podremos usar hasta que la transformemos a un valor numerico.

Para esto podemos usar label encoding o one-hot-encoding, veamos los efectos de cada una

### 3.1 one hot encoding

Lo que hace one hot encoding es crear una columna por cada valor diferente en la columna original, en este caso, usando la columna "color" creara dos nuevas columnas, una para rojo y otra para blanco.

Cada observación (fila) tendra un unico 1 en estas dos columnas, indicando de que color es, por ejemplo, si la fila 1 tiene un 1 en la columna "color_es_red", esto indica que este vino es rojo.

```python
df_dum = pd.get_dummies(df, columns=['color'], prefix=['color_es'])
df_dum.dtypes
```

```python
df_dum.head()
```

Veamos ahora la correlación con nuestra variable objetivo

```python
df_dum.corr()['quality'].nlargest(14)
```

Teniendo estas nuevas columnas, podemos ver su efecto en nuestra variable objetivo y usarlas al crear nuestro modelo

### 3.2 label encoding

A diferencia de one-hot-encoding, label encoding solo crea una columna nueva, donde a cada valor diferente de la columna original (color) le asigna un numero, por ejemplo 0= rojo y 1= blanco

```python
from sklearn.preprocessing import LabelEncoder
# creamos nuestro label encoder
labelencoder = LabelEncoder()
df['color_label'] = labelencoder.fit_transform(df['color'])
```

```python
df.head()
```

```python
df.tail()
```

```python
# eliminar columa color
df.drop('color', axis=1, inplace=True)
```

Ahora que tenemos un valor numerico, veamos de nuevo las correlaciones


```python
df.corr()['quality'].nlargest(14)
```

## 4 Conclusión

La selección de One-hot o label encoding tiene su impacto en el data set, y funcionan de manera diferentes, pero ¿Cómo sabemos cuál usar?

Primero, en el caso de one hot encoding, tengamos en cuenta que este va a crear tantas columnas como categorías que tenga la variable original. Por lo tanto, si tenemos muchas categorías, no es la mejor idea usar este algoritmo ya que tendrá un gran costo de memoria. Además, se recomienda usar este algoritmo cuando tenemos categorías nominales.

En el caso de label encoding, este algoritmo solo crea una columna así que el costo de memoria no es muy alto, y se recomienda usarlo cuando tenemos categorías ordinales.
